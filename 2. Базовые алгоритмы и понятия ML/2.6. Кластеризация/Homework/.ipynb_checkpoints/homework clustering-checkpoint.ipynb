```python
def unique_rows(a):
    a = np.ascontiguousarray(a)
    unique_a = np.unique(a.view([('', a.dtype)]*a.shape[1]))
    return unique_a.view(a.dtype).reshape((unique_a.shape[0], a.shape[1]))
```

```python
X_scaled = unique_rows(data_matrix)
```

```python
X_scaled[:5]
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, Birch, MeanShift,AffinityPropagation
from sklearn.metrics import adjusted_rand_score, silhouette_score
from sklearn.preprocessing import StandardScaler
```

### Определяем количество кластеров

```python
scaler = StandardScaler()
```

```python
k_inertia = []
ks = range(1,25)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(X_scaled, )
    k_inertia.append([clf_kmeans.inertia_])
```

### Начнем разбиение на кластеры:

```python
algorithms = [
    AffinityPropagation(damping=0.9),
    KMeans(n_clusters=11),
    AgglomerativeClustering(n_clusters=11),
    DBSCAN(eps=0.018),
    Birch(threshold = 0.0001 ,n_clusters = 11),
    MeanShift()
]
```

```python
def get_descr(algo_name, y_pred):
    predd = pd.DataFrame(y_pred)
    predd = predd.drop_duplicates().copy()
    Cluster_count = len(predd)
    return "{}\nCluster_count {:.0f}".format(
        algo_name,
        Cluster_count
    )
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
def print_graph(algorithms):
    for algorithm in algorithms:
        # кластеризуем и выводим картинку
        clusters = algorithm.fit_predict(X_scaled)
        plt.figure()
        plt.title(get_descr(algorithm.__class__.__name__,clusters))
        plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='autumn', s=60);  
        # если есть центры кластеров - выведем их
        if algorithm.__class__.__name__ in {'KMeans', 'AffinityPropagation','MeanShift'}:
            centers = algorithm.cluster_centers_
            plt.scatter(centers[:, 0], centers[:, 1], s=50)            
```

```python
print_graph(algorithms)
```
