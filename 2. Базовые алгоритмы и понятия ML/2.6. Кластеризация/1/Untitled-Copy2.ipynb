```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import numpy as np
```

```python
from sklearn.datasets import make_moons
from sklearn.datasets import make_blobs
```

```python
X, y = make_moons(n_samples=2000, noise=0.05, random_state=0)
```

```python
X.shape
```

```python
y.shape
```

```python
plt.scatter(X[:,0],
            X[:,1],
           c=y,
           cmap = 'autumn',
           s = 60);
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
```

```python
#from sklearn.metrics import 
```

```python
algos = [
    KMeans(n_clusters = 2),
    AgglomerativeClustering(n_clusters = 2),
    DBSCAN(eps=5)
]
```

```python
i = 0
for algo in algos:
    i = i + 1
    pred = algo.fit_predict(X)
    plt.figure()
    plt.title(algo.__class__.__name__)
    plt.scatter(X[:,0],
                X[:,1],
                c=pred,
               cmap = 'autumn',
               s = 60);
    if i < 2:
        plt.scatter(algo.cluster_centers_[:,0],
            algo.cluster_centers_[:,1],
            c = algo.cluster_centers_[0,:],
            s = 100)    
```

```python
X, y = make_blobs(n_samples=200, 
                  random_state=0, centers=2,cluster_std = 1.4)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans_pred = kmeans.fit_predict(X)
```

```python
plt.scatter(X[:,0],
            X[:,1],
           c=kmeans_pred,
           cmap = 'autumn',
           s = 60);
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            s = 100)
```
